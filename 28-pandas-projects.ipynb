```python
# Chipotle
chipo = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv")
```

```python
# See the first 10 entries
print(chipo.head(10))
```

```python
# What is the number of observations in the dataset?
print(len(chipo))
```

```python
# What is the number of columns in the dataset?
print(chipo.shape[1])
```

```python
# What is the number of columns in the dataset?
print(chipo.columns)
```

```python
# How is the dataset indexed?
print(chipo.index)
```

```python
# Which was the most-ordered item? 
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print(most_ordered_item)
```

```python
# For the most-ordered item, how many items were ordered?
item_counts = chipo['item_name'].value_counts()
quantity_ordered = item_counts.max()
print(quantity_ordered)
```

```python
# What was the most ordered item in the choice_description column?
print(chipo['choice_description'].value_counts().idxmax())
```

```python
# How many items were orderd in total?
print(len(chipo['item_name']))
```

```python
# Turn the item price into a float
chipo['item_price'] = chipo['item_price'].str[1:].astype(float)
```

```python
# Check the item price type
print(chipo['item_price'].dtypes)
```

```python
# Create a lambda function and change the type of item price
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x))
```

```python
# Check the item price type
print(chipo['item_price'].dtypes)
```

```python
# How much was the revenue for the period in the dataset?
chipo['revenue'] = chipo['item_price'] * chipo['item_price']
print("{:.3f}".format(chipo['revenue'].sum()))
```

```python
# How many orders were made in the period?
print(chipo['order_id'].nunique())
```

```python
# What is the average revenue amount per order?
average_revenue_per_order = chipo['revenue'].sum() / chipo['order_id'].nunique()
print("{:.3f}".format(average_revenue_per_order))
```

```python
# How many different items are sold?
print(chipo['item_name'].unique())
```
